```python
var_a = np.array(vals).var(ddof=1)
var_b = np.array(vals1).var(ddof=1)
N = len(vals)
s = np.sqrt((var_a + var_b)/2)
t = (np.mean(vals) - np.mean(vals1))/(s*np.sqrt(2/N))
df = 12

#p-value after comparison with the t 
scipy.stats.t.cdf(t, df=df)
```
