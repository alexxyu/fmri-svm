```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
inner_accs64 = pd.read_csv('output/inner_accs64.csv', sep='\t', index_col=0)

plt.figure(figsize=(10,10))
ax = sns.heatmap(inner_accs64.fillna(value=0), annot=True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
ax.set_title('SVM Inner Subject Accuracy w/ IP Data (No scramble)')
ax.set_ylabel('Outer Subject')
ax.set_xlabel('Inner Subject')
```

```python
outer_accs64 = pd.read_csv('output/outer_accs64.csv', sep='\t', index_col=0)

plt.figure(figsize=(10,10))
ax = sns.heatmap(outer_accs64.fillna(value=0), annot=True)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
ax.set_title('SVM Outer Subject Accuracy w/ IP Data (No scramble)')
ax.set_ylabel('Outer Subject')
ax.set_xlabel('Inner Subject')
```

```python
vals = []
for column in outer_accs64:
    vals.extend(outer_accs64[column].tolist())
vals = [x for x in vals if str(x) != 'nan']

plt.hist(vals, bins=20)
plt.xlabel('Accuracy')
plt.ylabel('Instances')
plt.title('Distribution of Outer Subject Testing Accuracies')
plt.savefig('output/outer_accs_hist.png')
```
